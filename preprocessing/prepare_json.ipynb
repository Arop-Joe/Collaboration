```python
import pandas as pd
import json
```

```python
df = pd.read_table('pantheon.tsv', sep='\t')
df.head()
```

```python
variables = ['en_curid', 'name', 'countryName', 'occupation']
```

```python
df[variables].head()
```

```python
x = df[variables]
x.rename(columns = {'en_curid': 'id', 'countryName': 'description', 'occupation': 'category_type'}, inplace=True)
x.head()
```

```python
data = []
# open a json file and save json data
with open('data.json', 'w') as outfile:
    for i in x.index:
        data.append(json.loads(x.loc[i].to_json()))
    json.dump(data, outfile)
```
